```python
import cv2, pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
per_dic_Water = {}
```

```python
img_2000 = cv2.imread("2000.tif")
```

```python
grayscale_image_2000 = img_2000[:, :, 0]

# Convert to binary by setting all non-zero values to 1
binary_image_2000 = (grayscale_image_2000 > 0).astype(int)

plt.imshow(binary_image_2000,  cmap='gray')  # 'gray' colormap to show it in grayscale
plt.colorbar()  # Optional: to show the color bar
plt.title('2000 Binary Image')
plt.show()
per_water_2000 = np.sum(binary_image_2000)/(binary_image_2000.shape[0]*binary_image_2000.shape[1])
per_dic_Water["Y-2000"] = per_water_2000*100
print("The Percentage of Water in 2000: "+str(np.round(per_water_2000*100,2))+"%")
```

```python
years = list(shrinkage_dict.keys())
percentages = list(shrinkage_dict.values())

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(years, percentages, marker='o', linestyle='-', color='b')
plt.title('Shrinkage of the Aral Sea Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage of Water Area (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
